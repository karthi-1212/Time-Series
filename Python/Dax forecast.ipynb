```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pmdarima as pm
from pmdarima.model_selection import train_test_split
```

```python
# Load the BJsales dataset
bj_sales_data = pm.datasets.load_wineind()

# Split into training and testing sets
train, test = train_test_split(bj_sales_data, train_size=0.8)  

# Fit ARIMA model automatically
model_bj = pm.auto_arima(train, seasonal=False, stepwise=True, suppress_warnings=True)

# Forecast the next 10 values
forecasted_values_bj = model_bj.predict(n_periods=10)
print("BJsales Forecasts (Next 10):\n", forecasted_values_bj)

# Plotting
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # Create a subplot for the original data
plt.plot(bj_sales_data)
plt.title('BJsales Data (Without Forecast)')

plt.subplot(1, 2, 2)  # Create a subplot for the forecast
plt.plot(bj_sales_data)
plt.plot(np.arange(len(bj_sales_data), len(bj_sales_data) + 10), forecasted_values_bj, linestyle='dashed', color='red')
plt.title('BJsales Data with 10-Step Forecast')
plt.show()
```
