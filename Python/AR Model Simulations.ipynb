```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Simulate AR(1) Process
np.random.seed(42)
y = np.random.normal(0, 2, 250)  # 250 random normal values
y1 = np.zeros_like(y)  # Initialize the array
y1[0] = y[0]
for i in range(1, 250):
    y1[i] = 0.5 * y1[i-1] + y[i]

plt.figure(figsize=(10, 4))
plt.plot(y1)
plt.title('Simulated AR(1) Process')
plt.xlabel('Time')
plt.ylabel('Value')

plot_acf(y1, lags=40, title='ACF of AR(1) Process')
plt.show()
```

```python
# AR(1) Model with Phi = 0.7
y = np.random.normal(0, 2, 250)
y1 = np.zeros_like(y)
y1[0] = y[0]
for i in range(1, 250):
    y1[i] = 0.7 * y1[i-1] + y[i]

plt.figure(figsize=(10, 4))
plt.plot(y1)
plt.title('Simulated AR(1) Process (Phi = 0.7)')
plt.xlabel('Time')
plt.ylabel('Value')

plot_acf(y1, lags=40, title='ACF of AR(1) Process (Phi = 0.7)')
plt.show()
```

```python

# AR(2) Model
y = np.random.normal(0, 2, 250)
y1 = np.zeros_like(y)
y1[0] = y[0]
y1[1] = y[1]
for i in range(2, 250):
    y1[i] = 0.7 * y1[i-1] + 0.3 * y1[i-2] + y[i]


plt.figure(figsize=(10, 4))
plt.plot(y1)
plt.title('Simulated AR(2) Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

plot_acf(y1, lags=40, title='ACF of AR(2) Process')
plt.show()
```

```python
# Fitting and Forecasting AR(2) Model
# Create a DataFrame for easy handling (with a dummy date index)
y1_df = pd.DataFrame({'value': y1}, index=pd.date_range(start='1-1-2023', periods=len(y1), freq='D'))

# Fit AR(2) model 
model = ARIMA(y1_df, order=(2, 0, 0))  # (p, d, q) where p=2 for AR(2)
fit_model = model.fit()

# Forecast
forecast_values = fit_model.forecast(steps=10)  # Forecast 10 steps ahead
print("Forecasted values for AR(2) model:", forecast_values)


# Plot original data and forecast
plt.plot(y1_df, label='Original Data')
plt.plot(forecast_values, label='Forecast')
plt.title('AR(2) Model Forecast')
plt.legend()
plt.show()
```
